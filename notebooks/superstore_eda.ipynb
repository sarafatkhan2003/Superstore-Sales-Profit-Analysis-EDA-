**PART 2: Check for Missing Values**

```python
# Identify Missing Data
df.isnull().sum()
```

How do you handle missing numerical data?
- By imputing with mean/median, using model-based methods, or dropping if insignificant.

When should you drop rows instead of filling?
- When missing values are few, random, or the row is not critical.

What risks come with filling missing values?
- It can introduce bias, reduce variance, and distort patterns.

**PART 3: Fix Data Types**

```python
# Convert Order Date & Ship Date

df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df.info()
```

**PART 4: Feature Engineering**

What Is Feature Engineering?

Creating new useful columns from existing data to:

- Improve analysis
- Reveal patterns
- Answer business questions

```python
# Create Time-Based Features
df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month
df['Order Day'] = df['Order Date'].dt.day_name()
```

```python
df['Shipping Days'] = (df['Ship Date'] - df['Order Date']).dt.days
```

**PART 3: Univariate Analysis on Numerical Variables**

Sales Distribution
Why Analyze Sales?

- Understand order value behavior
- Detect skewness
- Identify extreme orders

```python
sns.histplot(df['Sales'], kde=True)
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.show()
```

**Insight**

- Sales are right-skewed, indicating most orders are low-value with few high-value transactions.

```python
# Profit Distribution
sns.boxplot(x=df['Profit'])
plt.title("Profit Distribution")
plt.show()
```

**Insight**

- While most orders are profitable, there are several loss-making outliers.

```python
# Quantity Distribution
sns.histplot(df['Quantity'], bins=10)
plt.title("Quantity Distribution")
plt.show()
```

**insight**
- The histogram shows that most transactions involve lower quantities, with fewer high-quantity purchases, indicating typical small-volume buying behavior with occasional bulk orders.

**PART 5: Univariate Analysis on Categorical Variables**

```python
# Region Distribution
sns.countplot(y='Region', data=df)
plt.title("Orders by Region")
plt.show()
```

**SESSION 4: Bivariate & Multivariate Analysis**

**PART 1: What This Session Is About**

Analyzing two variables together to understand:

- Relationships
- Comparisons
- Patterns

**PART 2: Time-Based Analysis**

*Yearly Sales Trend*

```python
yearly_sales = df.groupby('Order Year')['Sales'].sum()

plt.figure(figsize=(8,4))
plt.plot(yearly_sales, marker='o')
plt.title("Yearly Sales Trend")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.show()
```

**Insight**

- Sales show a steady upward trend, indicating business growth over the years.

```python
# Monthly Seasonality
monthly_sales = df.groupby('Order Month')['Sales'].sum()

plt.plot(monthly_sales)
plt.title("Monthly Sales Seasonality")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

**Insight**

- Sales peak toward the end of the year, suggesting strong seasonal demand.

**PART 3: Category & Segment Analysis**

```python
# Sales by Category
sns.barplot(x = 'Category' , y = 'Sales', data = df)
plt.title('sales by category')
plt.show()
```

**Insight**

- Technology contributes the highest sales, while Furniture shows weaker performance.

```python
# Profit by Segment
sns.barplot(x='Segment', y='Profit', data=df)
plt.title("Profit by Customer Segment")
plt.show()
```

**Insight**

- Corporate segment generates higher profit per order compared to Consumer.

**PART 4: Region-wise Performance**

```python
# Profit by Region
region_profit = df.groupby('Region')['Profit'].sum()

region_profit.plot(kind='bar')
plt.title("Profit by Region")
plt.ylabel("Total Profit")
plt.show()
```

**Insight**

- The Central region shows losses and requires cost or pricing review.

**PART 5: Relationship Analysis (Correlation)**

```python
# Sales vs Quantity
sns.scatterplot(x = 'Quantity' , y = 'Sales', data = df)
plt.title('sales vs Quantity')
plt.show()
```

```python
# Correlation Heatmap
corr = df[['Sales','Quantity' , 'Discount','Profit']].corr()
sns.heatmap(corr , annot = True , cmap = 'coolwarm')
plt.title('correlation matrix')
plt.show()
```

- Sales and profit are positively correlated, while discount shows negative correlation with profit.

**PART 6: Multivariate Analysis**

```python
# Category + Region + Profit
sns.barplot(x='Region', y='Profit', hue='Category', data=df)
plt.title("Region-wise Profit by Category")
plt.show()
```

**Insight**

- Furniture category performs poorly in Central region, contributing to overall losses.

*Observation vs Insight*

- Observation = what you see
- Insight = what it means for the business + what to do

**PART 2: A Simple Framework to Write Insights**

Insight Formula

- What happened? (data fact)
- Why does it matter? (business impact)
- What should be done? (action / recommendation)

 Example

- What: Furniture category shows negative profit despite high sales.
- Why: High discounts and shipping costs are eroding margins.
- Action: Review pricing and discount strategy for Furniture.
